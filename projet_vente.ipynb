```python
import pandas as pd

#1- Chargement des données 
df = pd.read_csv("data/SampleSuperstore.csv")

#Afficher les données du csv
#df.head()

print(df.info())

print("----- Description sommaire des données -------")
print(df.describe())
```

```python
# 2- Nettoyer les données 
df = df.dropna()
```

```python
#3- Analyse des données
#3.1- Catégories les plus vendues

categories_plus_vendus = df.groupby("Category")["Sales"].sum().sort_values(ascending=False).head(10)
print("Catégories les plus vendues: \n", categories_plus_vendus)
```

```python
#3.2- Revenus par État
revenus_par_etat = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)
print("Revenus par État: \n", revenus_par_etat)
```

```python
#3.3- Catégories les plus rentables

categorie_plus_rentables = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print("Les catégories les plus rentables: \n", categorie_plus_rentables)
```

```python
#3.4- Meilleures sous-catégories
meilleures_sous_categories = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
print("Les meilleures sous-catégories: \n", meilleures_sous_categories)
```
